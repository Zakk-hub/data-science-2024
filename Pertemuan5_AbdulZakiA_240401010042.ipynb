Mempelajari cara menyajikan data dalam bentuk visual agar lebih mudah dipahami dan dianalisis.

```python
# import library
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

import seaborn as sns
```

```python
# Load Dataset
sns.set_theme(style='whitegrid')
df = sns.load_dataset('tips')
print(df.shape)
df.head()
```

```python
# buat dashboard
fig = plt.figure(figsize=(16,12))

fig.suptitle(
    'Dashboard Analisis Dataset Tips',
    fontsize=18,
    fontweight='bold'
)

gs = gridspec.GridSpec(
    2,
    2,
    figure=fig,
    hspace=0.4,
    wspace=0.35
)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

# Grafik 1
avg_by_day = df.groupby('day')['total_bill'].mean()
avg_by_day.plot(kind='bar', ax=ax1)
ax1.set_title('Rata-rata Total Bill per Hari')

# Grafik 2
sns.histplot(
    data=df,
    x='total_bill',
    hue='time',
    kde=True,
    ax=ax2
)
ax2.set_title('Distribusi Total Bill')

# Grafik 3
sns.boxplot(
    data=df,
    x='day',
    y='total_bill',
    hue='sex',
    ax=ax3
)
ax3.set_title('Distribusi Tagihan')

# Grafik 4
sns.scatterplot(
    data=df,
    x='total_bill',
    y='tip',
    hue='time',
    size='size',
    ax=ax4
)

sns.regplot(
    data=df,
    x='total_bill',
    y='tip',
    scatter=False,
    color='gray',
    ax=ax4
)

ax4.set_title('Hubungan Tagihan dan Tip')

plt.show()

```

# Bar Chart (Rata-rata Total Bill per Hari)
### What?
Grafik menunjukkan rata-rata total tagihan (total_bill) pelanggan pada setiap hari operasional restoran.
### So What?
Terlihat bahwa rata-rata total tagihan berbeda pada setiap hari. Hal ini menunjukkan adanya variasi pola pengeluaran pelanggan berdasarkan hari kunjungan.
### Now What?
Manajemen restoran dapat memanfaatkan informasi ini untuk menentukan strategi promosi, penjadwalan karyawan, dan pengelolaan stok pada hari-hari dengan rata-rata transaksi yang lebih tinggi.

# Histogram (Distribusi Total Bill)
### What?
Histogram menunjukkan distribusi nilai total_bill pelanggan beserta pola kepadatannya menggunakan KDE.
### So What?
Sebagian besar transaksi berada pada rentang tagihan rendah hingga menengah, sedangkan transaksi dengan nilai sangat besar jumlahnya lebih sedikit.
### Now What?
Restoran dapat memfokuskan strategi pelayanan dan promosi pada kelompok pelanggan dengan transaksi menengah yang merupakan mayoritas pelanggan.

# Boxplot (Distribusi Tagihan)
### What?
Boxplot menunjukkan distribusi total_bill berdasarkan hari dan jenis kelamin pelanggan.
### So What?
Terlihat adanya perbedaan median dan penyebaran nilai tagihan pada masing-masing kelompok. Selain itu terdapat beberapa outlier yang menunjukkan transaksi bernilai sangat tinggi.
### Now What?
Informasi ini dapat digunakan untuk memahami karakteristik pelanggan dan mengidentifikasi segmen pelanggan dengan kontribusi transaksi terbesar.

# Scatter Plot (Total Bill vs Tip)
### What?
Scatter plot menunjukkan hubungan antara total_bill dan tip yang diberikan pelanggan.
### So What?
Terlihat hubungan positif antara total_bill dan tip. Semakin besar nilai tagihan, semakin besar pula tip yang diberikan pelanggan.
### Now What?
Restoran dapat menggunakan informasi ini untuk memperkirakan pendapatan tip dan memahami perilaku pelanggan dalam memberikan tip berdasarkan nilai transaksi.
