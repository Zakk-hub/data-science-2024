# Langkah 1
Load & Inspect Dataset

```python
# Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
```

```python
# Load Dataset Iris
df = sns.load_dataset('iris')
print("Shape:", df.shape)

print("\nTipe Data:")
print(df.dtypes)

print("\n5 Data pertama:")
print(df.head())

print("\nStatistik Deskriptif:")
print(df.describe().round(3))
```

```python
# Cek Hasil
print(df.shape)
```

# Langkah 2
Statistik Deskriptif Lengkap

```python
for col_name in df.select_dtypes(include='number').columns:

    col = df[col_name]

    print(f'\n=== {col_name} ===')
    print(f'Mean     : {col.mean():.3f}')
    print(f'Median   : {col.median():.3f}')
    print(f'Std Dev  : {col.std():.3f}')
    print(f'Skewness : {col.skew():.3f}')
    print(f'Kurtosis : {col.kurt():.3f}')
```

## Interpretasi Petal Length

Berdasarkan hasil statistik deskriptif, variabel petal_length memiliki rata-rata sebesar 3.758 cm dan median sebesar 4.350 cm. Nilai skewness sebesar -0.275 menunjukkan distribusi data sedikit miring ke kiri (negative skew), namun masih cukup dekat dengan nol sehingga distribusinya relatif simetris.

Nilai kurtosis sebesar -1.402 menunjukkan distribusi lebih datar dibanding distribusi normal (platykurtic), sehingga data tidak terlalu terkonsentrasi pada nilai tengah. Standar deviasi sebesar 1.765 menunjukkan adanya variasi ukuran petal length antar spesies bunga Iris.

#Langkah 3
Analisis Distribusi Data Menggunakan Histogram dan KDE

```python
fig, ax = plt.subplots(figsize=(8,4))

sns.histplot(
    df['sepal_length'],
    kde=True,
    bins=20,
    ax=ax
)

ax.axvline(
    df['sepal_length'].mean(),
    color='red',
    linestyle='--',
    label=f"Mean={df['sepal_length'].mean():.2f}"
)

ax.axvline(
    df['sepal_length'].median(),
    color='orange',
    linestyle='--',
    label=f"Median={df['sepal_length'].median():.2f}"
)

ax.set_title('Distribusi Sepal Length')
ax.legend()

plt.show()
```

## Interprestasi Histogram
Histogram menunjukkan bahwa distribusi sepal_length relatif mendekati distribusi normal. Hal ini terlihat dari posisi mean dan median yang berdekatan. Sebagian besar data terkonsentrasi pada rentang nilai tengah dan tidak terlihat adanya outlier yang sangat ekstrem.

# Langkah 4
Boxplot dan Violin Plot Petal Length per Spesies

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))

sns.boxplot(
    data=df,
    x='species',
    y='petal_length',
    palette='Set2',
    ax=axes[0]
)

axes[0].set_title('Boxplot Petal Length')

sns.violinplot(
    data=df,
    x='species',
    y='petal_length',
    palette='Set2',
    inner='box',
    ax=axes[1]
)

axes[1].set_title('Violin Plot Petal Length')

plt.tight_layout()
plt.show()
```

## Interpretasi Boxplot

Berdasarkan boxplot dan violin plot, spesies Virginica memiliki nilai petal_length terbesar dibandingkan spesies lainnya. Sebaliknya, Setosa memiliki nilai petal_length terkecil.

Penyebaran data pada Virginica terlihat lebih besar dibandingkan Setosa, yang menunjukkan variasi ukuran petal yang lebih tinggi. Versicolor berada di antara keduanya dengan ukuran petal sedang.

# Langkah 5
Analisis Korelasi Pearson

```python
corr = df.drop('species', axis=1).corr(method='pearson')

print(corr.round(3))
```

```python
import numpy as np

mask = np.triu(np.ones(corr.shape, dtype=bool))

corr_masked = corr.where(~mask)

max_pair = corr_masked.stack().idxmax()

print(f'Korelasi tertinggi: {max_pair} = {corr.loc[max_pair]:.3f}')
```

## Interpretasi Korelasi

Hasil analisis menunjukkan bahwa pasangan variabel petal_length dan petal_width memiliki korelasi tertinggi sebesar 0.963. Nilai ini termasuk kategori sangat kuat dan positif.

Artinya, semakin panjang petal bunga, maka cenderung semakin lebar pula ukuran petalnya. Hubungan ini hampir linear dan menunjukkan keterkaitan yang sangat erat antara kedua variabel tersebut.

Namun perlu diingat bahwa korelasi tidak selalu menunjukkan hubungan sebab-akibat (kausalitas).

# Langkah 6
Scatter Plot dan Heatmap Korelasi

```python
fig, axes = plt.subplots(1, 2, figsize=(14,5))

sns.scatterplot(
    data=df,
    x='petal_length',
    y='petal_width',
    hue='species',
    palette='Set2',
    ax=axes[0]
)

sns.regplot(
    data=df,
    x='petal_length',
    y='petal_width',
    scatter=False,
    color='gray',
    ax=axes[0]
)

axes[0].set_title('Scatter: Petal Length vs Petal Width')

sns.heatmap(
    corr,
    annot=True,
    fmt='.2f',
    cmap='RdBu_r',
    vmin=-1,
    vmax=1,
    square=True,
    ax=axes[1]
)

axes[1].set_title('Heatmap Korelasi')

plt.tight_layout()
plt.show()
```

## Interpretasi Scatter Plot dan Heatmap

Scatter plot menunjukkan hubungan positif yang sangat kuat antara petal_length dan petal_width. Titik-titik data membentuk pola yang cenderung naik, menunjukkan bahwa semakin besar nilai petal_length maka semakin besar pula nilai petal_width.

Heatmap korelasi memperlihatkan bahwa sebagian besar variabel memiliki korelasi positif. Korelasi tertinggi terdapat pada pasangan petal_length dan petal_width dengan nilai sekitar 0.963 yang termasuk kategori sangat kuat.
