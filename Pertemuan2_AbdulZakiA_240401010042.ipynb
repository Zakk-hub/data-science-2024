```python
#Load Dataset Titanic
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)

df.head()
```

# Q1
Menampilkan Jumlah baris, kolom, dan nama kolom dataset Titanic

```python
#Kode Q1
print("Shape Dataset:")
print(df.shape)

print("\nDaftar Kolom:")
print(df.columns.tolist())
```

Dataset Titanic memiliki 891 baris dan 12 kolom yang berisi informasi penumpang seperti nama, jenis kelamin, usia, kelas tiket, tarif, dan status keselamatan.

# Q2
Mencari missing values.

```python
missing = df.isnull().sum()

persentase = (missing / len(df)) * 100

hasil_missing = pd.DataFrame({
    'Jumlah Missing': missing,
    'Persentase (%)': persentase
})

hasil_missing[hasil_missing['Jumlah Missing'] > 0]
```

Kolom Age, Cabin, dan Embarked memiliki data yang kosong (missing values). Kolom Cabin memiliki jumlah data kosong paling banyak sehingga perlu penanganan khusus sebelum digunakan untuk analisis lanjutan.

# Q3
Menghitung persentase penumpang yang selamat.

```python
persen_selamat = df["Survived"].mean() * 100

print(f"Persentase penumpang yang selamat: {persen_selamat:.2f}%")
```

Sekitar 38% penumpang Titanic berhasil selamat, sedangkan sisanya tidak selamat

# Q4
Mencari jumlah penumpang wanitas pada kelas 1 dan rata rata usianya.

```python
wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
jumlah = wanita_kelas1.shape[0]
rata_usia = wanita_kelas1['Age'].mean()

print("jumlah wanita kelas 1:", jumlah)
print("Rata-rata usia:", round(rata_usia, 2))
```

Penumpang wanita kelas 1 memiliki jumlah tertentu dan rata-rata usia yang menunjukkan kelompok penumpang kelas atas pada masa tersebut.

# Q5
Menghitung tingkat keselamatan berdasarkan kelas penumpang

```python
hasil = pd.DataFrame({
    "Persentase Selamat (%)":
    df.groupby("Pclass")["Survived"].mean() * 100
})

hasil
```

Penumpang kelas 1 memiliki tingkat keselamatan tertinggi dibandingkan kelas lainnya. Hal ini menunjukkan bahwa status sosial dan lokasi kabin berpengaruh terhadap peluang keselamatan.

# Q6
Analisis tingkat keselamatan berdasarkan jenis kelamin

```python
hasil_gender = pd.DataFrame({
    "Persentase Selamat (%)":
    df.groupby("Sex")["Survived"].mean() * 100 })

hasil_gender
```

Penumpang perempuan memiliki tingkat keselamatan yang jauh lebih tinggi dibandingkan laki-laki. Hal ini sesuai dengan kebijakan evakuasi Titanic yang mengutamakan perempuan dan anak-anak.
